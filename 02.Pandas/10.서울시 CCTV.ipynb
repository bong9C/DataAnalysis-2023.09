### 2.10 서울시 구별 CCTV 현황 분석
- 서울시 자치구 연도별 CCTV 운영현황
- 서울시 주민등록 인구(구별 통계)

    - cctv 갯수 : 2015년 ~ 2022년 계
    - 최근 증가율 : (2020 ~ 2022) / 총 댓수 * 100
    - 인구수 : 2023 1/4 분기 자료 
       종로구 중구 용산구

### 1.CCTV 데이터

```python
import numpy as np
import pandas as pd
```

```python
cctv = pd.read_csv('data/서울시 자치구 년도별 CCTV 설치 현황_230630기준.csv',encoding='EUC-KR', index_col='구분',thousands=',')

cctv
```

```python
cctv.drop(columns=['Unnamed: 0', '총계', '2023년'], inplace=True)
cctv.drop(['계'], inplace=True)
cctv.head()
```

```python
# 결측치가 있는지 확인 
cctv['2014년 이전'].isna().sum()
```

```python
cctv.isna().sum(), cctv.isna().sum()
```

```python
# 결측치를 0으로 대체하고, 타입을 정수로 변경

cctv = cctv.fillna(0).astype(int) # (0.0)으로 해도 됨
cctv
```

```python
cctv['CCTV댓수'] = cctv.sum(axis=1)
cctv.head(3)
```

```python
cctv['최근증가율'] = (cctv.loc[:,'2020년':'2022년'].sum(axis=1) / cctv['CCTV댓수'] * 100).round(2)
cctv.head(3)
```

```python
cctv.drop(columns= ['2014년 이전']+'2014년 2015년   2016년   2017년   2018년   2019년   2020년   2021년   2022년'.split(),
                     inplace=True)
cctv.head()
```

```python
#'중 구'사이의 Blank 없애기
cctv.index = cctv.index.str.replace(' ','')
cctv.index[1]
```

### 2. 서울시 구별 주민등록 데이터

```python
pop = pd.read_csv('data/서울시 구별 주민등록.csv',skiprows=1, thousands=',')
pop.head(5)
```

```python
pop = pop[['동별(2)','계 (명)','한국인 (명)','등록외국인 (명)','65세이상고령자 (명)']]
pop.head(3)
```

```python
pop.columns = ['구별','인구수','내국인','외국인','고령자']
pop.head(3)
```

```python
# 첫번째 행 지우기 
pop.drop([0], inplace=True)
pop.head(3)
```

```python
# 구별을 인덱스로 만들기 
pop.set_index('구별', inplace=True)
pop.head()
```

```python
# 외국인 비율
pop['외국인 비율'] = (pop.외국인 / pop['인구수'] * 100).round(2)
pop['고령자 비율'] = (pop['고령자'] / pop['인구수'] * 100).round(2)
pop.head()
```

### 3. CCTV 데이터와 POP 데이터 합치기 

```python
so = pd.concat([cctv, pop], axis=1)
so.head()
```

```python
df = pd.merge(cctv, pop, left_index=True, right_index=True)
df.head()
```

```python
df.shape
```

```python
df.index.name = '구별'
df.head()
```

-  서울시 구별 CCTV 인구 현황

```python
# 데이터 파일 만들기
df.to_csv('data/서울시 구별 CCTV 인구 현황.csv')
```
