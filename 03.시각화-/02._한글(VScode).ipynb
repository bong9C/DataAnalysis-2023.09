- 곡선 그리기

```python
X = np.linspace(0,2*np.pi, 200)     # np.linspace 라인을 몇등분 할건지. 
Ysin, Ycos = np.sin(X), np.cos(X)
plt.plot(X,Ysin), plt.plot(X, Ycos)
plt.title(' Sine, Cosine 그래프'), plt.grid()
plt.show()
```

- Tick 설정

```python
plt.plot(X,Ysin), plt.plot(X, Ycos)
plt.title('Tick 설정'), plt.grid()
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi])
plt.yticks([-1, -0.707, 0, 0.707, 1])
plt.show()
```

```python
plt.plot(X,Ysin), plt.plot(X, Ycos)
plt.title('LaTeX, 문자열로 tick label 정의'), plt.grid()
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           [r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'])    # LaTeX
plt.yticks([-1, -0.707, 0, 0.707, 1],
           [r'$-1$', r'$-\sqrt{2}/2$', r'$0$', r'$\sqrt{2}/2$', r'$1$'])
plt.show()
```

- 법례(legend) 표시

```python
plt.plot(X,Ysin, label='사인')
plt.plot(X, Ycos, label='코사인')
plt.title('범례표시'), plt.grid()
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           [r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'])    # LaTeX
plt.yticks([-1, -0.707, 0, 0.707, 1],
           [r'$-1$', r'$-\sqrt{2}/2$', r'$0$', r'$\sqrt{2}/2$', r'$1$'])
plt.legend(loc='upper center')      # 파라메터를 생략하면 적절한 위치에 표시해줌
plt.show()
```

- X축, Y축 Label

```python
plt.plot(X,Ysin, label='사인')
plt.plot(X, Ycos, label='코사인')
plt.title('X축, Y축 Lable 표시'), plt.grid()
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           [r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'])    # LaTeX
plt.yticks([-1, -0.707, 0, 0.707, 1],
           [r'$-1$', r'$-\sqrt{2}/2$', r'$0$', r'$\sqrt{2}/2$', r'$1$'])
plt.xlabel('시간'), plt.ylabel('진폭')
plt.legend()                       # 파라메터를 생략하면 적절한 위치에 표시해줌
plt.show()
```

- 그림 크기, 제목 크기 지정

```python
plt.figure(figsize=(10,6))
plt.plot(X,Ysin, label='사인')
plt.plot(X, Ycos, label='코사인')
plt.title('X축, Y축 Lable 표시', fontsize=20)
plt.xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
           [r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'])    # LaTeX
plt.yticks([-1, -0.707, 0, 0.707, 1],
           [r'$-1$', r'$-\sqrt{2}/2$', r'$0$', r'$\sqrt{2}/2$', r'$1$'])
plt.xlabel('시간'), plt.ylabel('진폭')
plt.legend(), plt.grid()                       # 파라메터를 생략하면 적절한 위치에 표시해줌
plt.show()
```
