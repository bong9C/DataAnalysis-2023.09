## 서울시 CCTV 인구 시각화

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('../02.Pandas/data/서울시 구별 CCTV 인구 현황.csv', index_col='구별')
df.head()
```

1. 자치구별 CCTV 설치 댓수

```python
df.CCTV댓수.plot(kind= 'barh')
plt.title('서울시 구별 CCTV 설치 댓수');
```

```python
# CCTV 댓수로 정렬하고, plt.bath() 로 그리기
sc = df['CCTV댓수'].sort_values()
plt.barh(sc.index, sc.values)
plt.title('서울시 구별 CCTV 설치 갯수 정렬 완료');
```

```python
# Seabron으로 그리기
sns.barplot(data=df, y=df.index, x='CCTV댓수') 
plt.title('서울시 구별 CCTV 설치 댓수');
```

```python
scr = df['CCTV댓수'].sort_values(ascending=False)
sns.barplot(y=sc.index, x=scr.values)
plt.title('서울시 구별 CCTV 설치 댓수');
```

##### 2. 구별 인당 CCTV 댓수 그리기 - plt.barh() 로 그리기


```python
df['인당 CCTV댓수'] = df['CCTV댓수'] / df['인구수']
df
```

```python
#이렇게도 할 수 있음 
df['CCTV댓수'] / df['인구수'].sort_values()   #sort_values() 를 하면 바로 인당 CCTV 댓수만 나옴
```

```python
sc = df['인당 CCTV댓수'].sort_values()
plt.barh(sc.index, sc.values)
plt.title('서울시 구별 인당 CCTV 설치 갯수 정렬 완료');
```

```python
# 이렇게 바로 할 수도 있음 
s = (df['CCTV댓수'] / df['인구수']).sort_values()   #sort_values() 를 하면 바로 인당 CCTV 댓수만 나옴
plt.barh(s.index, s.values)
plt.title('구별 인당 CCTV 댓수');
```
